```
# onehot encoding for cp_type
categ = feature.select('cp_type').distinct().rdd.flatMap(lambda x:x).collect()
exprs = [F.when(F.col('cp_type') == cat,1).otherwise(0)\
            .alias(str(cat)) for cat in categ]
df1 = feature.select(['sig_id']+exprs)
```

```
# onehot encoding for cp_dose
categ = feature.select('cp_dose').distinct().rdd.flatMap(lambda x:x).collect()
exprs = [F.when(F.col('cp_dose') == cat,1).otherwise(0)\
            .alias(str(cat)) for cat in categ]
df2 = feature.select(['sig_id']+exprs)
```

```
# ordinal encoding for cp_time
df3 = feature.withColumn("cp_time", when(col("cp_time") == 24, 1)
                                 .when(col("cp_time") == 48, 2)
                                 .when(col("cp_time") == 72, 3)
                                 .otherwise(0)).select("sig_id","cp_time")
```

```
df3.show(5)
```

```
# merge the dataframe and remove the redundant
df_drop = feature.drop(*["cp_type","cp_dose","cp_time"])
df_join1 = df_drop.join(df1, on = ["sig_id"], how = "right")
df_join2 = df_join1.join(df2, on = ["sig_id"], how = "right")
df_feature = df_join2.join(df3, on = ["sig_id"], how = "right")
```
